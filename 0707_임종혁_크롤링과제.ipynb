```python
import requests

url = 'https://www.musinsa.com/app/codimap/lists?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=comment_cnt&page=1'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

res = requests.get(url, headers=headers)

print(res)  # 응답 객체
print(res.text)  # 응답 내용
```

```python
res.status_code # HTTP 상태 코드
```

```python
import pandas as pd

res = requests.get(url, headers=headers)

if res.status_code == 200:
    soup = BeautifulSoup(res.text, 'html.parser')

    data = []
    items = soup.find_all('li', class_='style-list-item')
    for item in items:
        category = item.find(class_='style-list-information__text').text.strip()
        title = item.find(class_='style-list-information__title').text.strip()
        date = item.find(class_='post-information__text').text.strip()

        # 조회수 값 추출
        views_span = item.find(class_='post-information').find_all(class_='post-information__text')[1]
        views = views_span.text.strip().replace('조회', '') if views_span else ''

        # 댓글 수 추출
        comment_span = item.find(class_='post-information').find_all(class_='post-information__text')[2]
        comment_numbers = comment_span.text.strip().replace('댓글', '') if comment_span else ''

        data.append([category, title, date, views, comment_numbers])

    df = pd.DataFrame(data, columns=['codimap_category', 'codimap_title', 'codimap_date', 'views', 'comment_numbers'])
    df.to_csv('0707_임종혁_크롤링과제.csv', index=False, encoding='cp949')

    from google.colab import files
    files.download('0707_임종혁_크롤링과제.csv')
    print('CSV 파일로 저장되었습니다.')

else:
    print('GET 요청 실패:', res.status_code)
```
