```python
df1[['승차_정류장ARS', '하차_정류장ARS']] = df1[['승차_정류장ARS', '하차_정류장ARS']].apply(lambda x: x.astype(int))
df1
```

```python
df2_sample = df2[['ARS_ID', 'X좌표', 'Y좌표']]
df2_sample
```

```python
# 승차_정류장ARS에 해당하는 X좌표와 Y좌표를 df2_sample에서 가져와서 df1에 추가합니다.
df1_merged = pd.merge(df1, df2_sample, how='inner', left_on='승차_정류장ARS', right_on='ARS_ID')
df1_merged = df1_merged.rename(columns={'X좌표': '승차_정류장_X좌표', 'Y좌표': '승차_정류장_Y좌표'})

# 하차_정류장ARS에 해당하는 X좌표와 Y좌표를 df2_sample에서 가져와서 df1에 추가합니다.
df1_merged = pd.merge(df1_merged, df2_sample, how='inner', left_on='하차_정류장ARS', right_on='ARS_ID')
df1_merged = df1_merged.rename(columns={'X좌표': '하차_정류장_X좌표', 'Y좌표': '하차_정류장_Y좌표'})

df1=df1.dropna(subset=['승차_정류장_X좌표', '승차_정류장_Y좌표', '하차_정류장_X좌표', '하차_정류장_Y좌표'])

df1
```

```python
def dis(lat1, lon1, lat2, lon2):
    distance = ((lat2 - lat1) ** 2 + (lon2 - lon1) ** 2) ** 0.5 * 133.33
    return distance

# 승차정류장과 하차정류장의 좌표를 이용하여 거리를 계산하여 '거리' 열을 추가합니다.
df1['거리'] = dis(df1['승차_정류장_Y좌표'], df1['승차_정류장_X좌표'], df1['하차_정류장_Y좌표'], df1['하차_정류장_X좌표'])
df1
```

```python
# 거리에 따라 내림차순으로 df1을 정렬합니다.
df1 = df1.sort_values(by='거리', ascending=False)

# 인덱스를 초기화하고 인덱스 열을 제거합니다.
df1 = df1.reset_index(drop=True)
df1
```

```python
route_counts = df1.groupby('노선명').size().reset_index(name='size')

# 각 경로의 거리에 대한 최소값, 최대값, 평균값을 계산합니다.
route_stats = df1.groupby('노선명')['거리'].agg(['min', 'max', 'mean']).reset_index()

# '노선명'을 기준으로 route_counts와 route_stats를 병합합니다.
df1 = pd.merge(route_counts, route_stats, on='노선명')

df1
```
